```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load your dataset (replace 'your_dataset.csv' with your actual dataset file)
xlsx_file = '/content/Autograde_Dataset.xlsx'
df = pd.read_excel(xlsx_file)
df.dropna(inplace=True)
print(df.columns)

# Assuming 'Homework', 'Task', etc., are your text features
text_features = df[['Task', 'Task_Solution_Code', 'Task_Solution_Report', 'Task_Solution_Expectations']]

# Target variable
target = df['Task_Solution_Result_Points']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(text_features, target, test_size=0.1, random_state=42)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train.apply(lambda x: ' '.join(x), axis=1))
X_test_tfidf = tfidf_vectorizer.transform(X_test.apply(lambda x: ' '.join(x), axis=1))

# Linear Regression Model
model = LinearRegression()
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R^2 Score: {r2}')

cvmodel = RandomForestRegressor()
cvmodel.fit(X_train_tfidf, y_train)  # Fit the model on the training set
y_pred_cv = cvmodel.predict(X_test_tfidf)  # Predict on the test set

# Evaluate the model
mse_cv = mean_squared_error(y_test, y_pred_cv)
print(f'Mean Squared Error (Random Forest): {mse_cv}')
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# Load your dataset (replace 'your_dataset.csv' with your actual dataset file)
xlsx_file = '/content/Autograde_Dataset.xlsx'
df = pd.read_excel(xlsx_file)
df.dropna(inplace=True)
print(df.columns)

# Assuming 'Homework', 'Task', etc., are your text features
text_features = df[['Task', 'Task_Solution_Code', 'Task_Solution_Report', 'Task_Solution_Expectations']]

# Target variable
target = df['Task_Solution_Result_Points']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(text_features, target, test_size=0.1, random_state=42)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train.apply(lambda x: ' '.join(x), axis=1))
X_test_tfidf = tfidf_vectorizer.transform(X_test.apply(lambda x: ' '.join(x), axis=1))

# Linear Regression Model
linear_model = LinearRegression()
linear_model.fit(X_train_tfidf, y_train)
y_pred_linear = linear_model.predict(X_test_tfidf)

# Evaluate the Linear Regression model
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print(f'Linear Regression - Mean Squared Error: {mse_linear}')
print(f'Linear Regression - R^2 Score: {r2_linear}')

# Logistic Regression Model
# Assuming 'Task_Solution_Result_Points' is a binary classification target
logistic_model = LogisticRegression()
logistic_model.fit(X_train_tfidf, y_train)
y_pred_logistic = logistic_model.predict(X_test_tfidf)

# Evaluate the Logistic Regression model
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(f'Logistic Regression - Accuracy: {accuracy_logistic}')
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load your dataset (replace 'your_dataset.csv' with your actual dataset file)
xlsx_file = '/content/Autograde_Dataset.xlsx'
df = pd.read_excel(xlsx_file)
df.dropna(inplace=True)
print(df.columns)


text_features = df[['Task', 'Task_Solution_Code', 'Task_Solution_Report', 'Task_Solution_Expectations']]

# Target variable
target = df['Task_Solution_Result_Points']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(text_features, target, test_size=0.1, random_state=42)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train.apply(lambda x: ' '.join(x), axis=1))
X_test_tfidf = tfidf_vectorizer.transform(X_test.apply(lambda x: ' '.join(x), axis=1))

# K-Nearest Neighbors Regressor Model
knn_model = KNeighborsRegressor(n_neighbors=3)  # You can adjust the number of neighbors as needed
knn_model.fit(X_train_tfidf, y_train)
y_pred_knn = knn_model.predict(X_test_tfidf)

# Evaluate the K-Nearest Neighbors Regressor model
mse_knn = mean_squared_error(y_test, y_pred_knn)
r2_knn = r2_score(y_test, y_pred_knn)
print(f'K-Nearest Neighbors - Mean Squared Error: {mse_knn}')
print(f'K-Nearest Neighbors - R^2 Score: {r2_knn}')

# Recurrent Neural Network (RNN) Model
# Note: This is a simple example. Depending on your data, you might need a more complex RNN architecture.
model_rnn = Sequential()
model_rnn.add(Embedding(input_dim=5000, output_dim=32, input_length=X_train_tfidf.shape[1]))
model_rnn.add(LSTM(100))
model_rnn.add(Dense(1, activation='linear'))
model_rnn.compile(optimizer='adam', loss='mean_squared_error')

# Convert text data to sequences
X_train_sequences = tf.keras.preprocessing.sequence.pad_sequences(X_train_tfidf.toarray(), maxlen=X_train_tfidf.shape[1])
X_test_sequences = tf.keras.preprocessing.sequence.pad_sequences(X_test_tfidf.toarray(), maxlen=X_train_tfidf.shape[1])

# Train the RNN model
model_rnn.fit(X_train_sequences, y_train, epochs=10, batch_size=32, validation_split=0.1)

# Predict using the RNN model
y_pred_rnn = model_rnn.predict(X_test_sequences)

# Evaluate the RNN model
mse_rnn = mean_squared_error(y_test, y_pred_rnn)
r2_rnn = r2_score(y_test, y_pred_rnn)
print(f'Recurrent Neural Network - Mean Squared Error: {mse_rnn}')
print(f'Recurrent Neural Network - R^2 Score: {r2_rnn}')
```
